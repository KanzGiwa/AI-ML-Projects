```python
import pandas as pd
import numpy as np
np.set_printoptions(formatter={'float_kind':"{:3.2f}".format})
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from confint import classification_confint

df = pd.read_csv(url)
df = df.drop(['ID'], axis = 1)
x = df.drop(['treat'], axis = 1)
y = df['treat']

actual_y = df['treat']
```

```python
# SVM model
model1 = SVC(kernel = 'linear', max_iter=10000)
model2 = SVC(kernel = 'poly', coef0 = 1.0, max_iter=10000)
model3 = SVC(kernel = 'rbf', max_iter=10000)
model4 = SVC(kernel = 'sigmoid', max_iter=10000)
```

```python
C = [10**n for n in range(-5,6)]
```

```python
model = SVC(max_iter=10000)
```

```python
model1.fit(x, actual_y)
model2.fit(x, actual_y)
model3.fit(x, actual_y)
model4.fit(x, actual_y)
```

```python
linearPrediction = model1.predict(x)
polyPrediction = model2.predict(x)
rbfPrediction = model3.predict(x)
sigmoidPrediction = model4.predict(x)
```

```python
acc = accuracy_score(actual_y, linearPrediction)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
acc = accuracy_score(actual_y, polyPrediction)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
acc = accuracy_score(actual_y, rbfPrediction)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
acc = accuracy_score(actual_y, sigmoidPrediction)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
# grid search
param_grid = [
  {'C': C, 'kernel': ['linear']},
  {'C': C, 'degree': [2, 3, 4, 5, 6], 'coef0': [1.0], 'kernel': ['poly']},
  {'C': C, 'gamma': ['scale', 'auto'], 'kernel': ['rbf']},
  {'C': C, 'coef0': [0.5, 1.0, 1.5], 'kernel': ['sigmoid']},
 ]
grid = GridSearchCV(model, param_grid, cv=5)
grid.fit(x, actual_y)
print("Grid Search: best parameters: {}".format(grid.best_params_))
```

The best parameters are {'C': 10000, 'coef0': 1.0, 'degree': 6, 'kernel': 'poly'}

```python
# evaluate the best model
best_model = grid.best_estimator_
predict_y = best_model.predict(x)
acc = accuracy_score(actual_y, predict_y)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
model5 = SVC(kernel = 'sigmoid', coef0 = 0.5, gamma = 'auto', max_iter=10000)
model5.fit(x, actual_y)

predict_y = model5.predict(x)
acc = accuracy_score(actual_y, predict_y)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
model6 = SVC(kernel = 'sigmoid', coef0 = 0.5, gamma = 'scale', max_iter=10000)
model6.fit(x, actual_y)

predict_y = model6.predict(x)
acc = accuracy_score(actual_y, predict_y)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
model7 = SVC(kernel = 'poly', coef0 = 0.5, gamma = 'auto', max_iter=10000)
model7.fit(x, actual_y)

predict_y = model7.predict(x)
acc = accuracy_score(actual_y, predict_y)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
model8 = SVC(kernel = 'poly', coef0 = 0.5, gamma = 'scale', max_iter=10000)
model8.fit(x, actual_y)

predict_y = model8.predict(x)
acc = accuracy_score(actual_y, predict_y)
lb,ub = classification_confint(acc,x.shape[0])
print("Accuracy: {:3.2f} ({:3.2f},{:3.2f})".format(acc,lb,ub))
```

```python
# grid search
param_grid = [
  {'C': C, 'degree': [2, 3, 4, 5, 6], 'coef0': [1.0], 'kernel': ['poly']},
  {'C': C, 'coef0': [0.5, 1.0, 1.5], 'kernel': ['sigmoid']},
 ]

grid = GridSearchCV(model, param_grid, cv=5)
grid.fit(x, actual_y)
print("Grid Search: best parameters: {}".format(grid.best_params_))
```

```python
# build the confusion matrix
labels = df['treat'].unique()
cm = confusion_matrix(actual_y, predict_y, labels=labels)
cm_df = pd.DataFrame(cm, index=labels, columns=labels)
print("Confusion Matrix:\n{}".format(cm_df))
```
