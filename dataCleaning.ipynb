Question:Which column has the maximum missing value?

Answer: The "Province State" is the column with the maximum missing value.

```python
percentage_missing = (df['Province_State'].isnull().sum() / len(df)) * 100
print(f"Percentage of missing values in 'Province_State': {percentage_missing:.2f}%")
```

Question: What % value is missing in that column?

Answer: Percentage of missing values in 'Province_State': 25.87%

Question:Should you keep this column or drop it? Why?

Answer: I should keep this column because the percentage of value sthat is missing isn't high and the column has 132 unique information which can be valuable information

```python
print(f"Unique values in 'Province_State': {df['Province_State'].nunique()}")
print(f"Unique values in 'Confirmed': {df['Confirmed'].nunique()}")
print(f"Unique values in 'Deaths': {df['Deaths'].nunique()}")
print(f"Unique values in 'Recoverd': {df['Recovered'].nunique()}")
print(f"Unique values in 'Active': {df['Active'].nunique()}")
```

Question: There are some columns with only a few data points missing. Should you drop these columns too? Why or why not?

Answer: I should keep the other columns that have a few missing data points because they have a lot of unique information and also they are significant columns that are important for identifying.

Question: How should you handle these specific data points(particular row)?

Answer: I will handle the specific data points that are missing by leaving them as they are.

Question: There are some data that has incorrect information like instead of a number they might have strings. How are you handling these data points(rows)?

Answer: If it has incorrect information in that row I will impute it with a mean value

```python
df['Province_State'] = pd.to_numeric(df['Province_State'], errors='coerce')

#fill NaN values with the mean
df['Province_State'].fillna(df['Province_State'].mean(), inplace=True)
df.Province_State[df.Province_State.isnull()]
```

```python
df['Confirmed'].fillna(df['Confirmed'].mean(), inplace=True)
df.Confirmed[df.Confirmed.isnull()]
```

```python
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')

#fill NaN values with the mean
df['Deaths'].fillna(df['Deaths'].mean(), inplace=True)
df.Deaths[df.Deaths.isnull()]
```

```python
df['Recovered'].fillna(df['Recovered'], inplace=True)
df.Recovered[df.Recovered.isnull()]
```

```python
df['Active'].fillna(df['Active'], inplace=True)
df.Active[df.Active.isnull()]
```

```python
top_10_countries = df.groupby('Country_Region')['Confirmed'].sum().nlargest(10)
print(top_10_countries)
```

```python
cleaned_df = df.copy()
print(cleaned_df.head())
cleaned_df.to_csv('cleaned_covid19_southamerica.csv', index=False)
```
